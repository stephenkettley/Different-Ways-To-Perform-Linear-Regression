```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

# 1. The Normal Equation

## Let's Generate Linear Data

```python
X = 2 * np.random.rand(100,1)
y = 4 + 3 * X +  np.random.randn(100,1)
```

```python
plt.scatter(X,y, c='g')
plt.title('Randomly Generated Linear Dataset', fontsize=20)
plt.xlabel('X1', fontsize=12)
plt.ylabel('y', fontsize=12)
```

## Let's Compute Theta Hat using the Normal Equation

```python
X_b = np.c_[np.ones((100,1)), X] # add x0 = 1 to each inverse
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y) #this calculates the value of theta that minimizes the MSE cost function
```

```python
theta_best
```

## Let's Make Predictions Using Theta Hat

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

## Let's Plot The Predictions

```python
plt.plot(X_new, y_predict, 'r-', label='Predictions')
plt.plot(X,y, 'b.')
plt.axis([0,2,0,15])
plt.title('Linear Regression Model Predictions', fontsize=22)
plt.xlabel('X1', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.legend()
plt.show()
```
